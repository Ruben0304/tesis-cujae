# Potencia de referencia del dataset

Este cuaderno calcula la potencia pico del sistema fotovoltaico 
presente en `Solar Power Plant Data.csv`, el mismo dataset que se usó para 
entrenar el modelo de producción. El objetivo es conocer la potencia 
máxima observada para poder escalar las predicciones a otros arreglos 
fotovoltaicos.


```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("datasets/clima paneles/Solar Power Plant Data.csv")
df = pd.read_csv(DATA_PATH)

print(f"Forma del dataset: {df.shape}")
df.head(10)
```

```python
production = df['SystemProduction']

reference_capacity_kw = production.max()
p999 = production.quantile(0.999)
p995 = production.quantile(0.995)
p99 = production.quantile(0.99)
mean_active = production[production > 0].mean()

print(f"Potencia máxima observada: {reference_capacity_kw:.2f} kW")
print(f"Percentil 99.9%: {p999:.2f} kW")
print(f"Percentil 99.5%: {p995:.2f} kW")
print(f"Percentil 99%: {p99:.2f} kW")
print(f"Promedio cuando hay generación: {mean_active:.2f} kW")

print(f"\nPara escalar a una potencia objetivo P, usa: prediccion * (P / {reference_capacity_kw:.2f})")
```
